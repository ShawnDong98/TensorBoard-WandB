## 1. Log a table

初始化运行，创建一个 `wandb.Table()`，然后将其记录到运行中。

```python
import wandb

run = wandb.init(project="table-test")
my_table = wandb.Table(columns=["a", "b"], data=[["a1", "b1"], ["a2", "b2"]])
run.log({"Table Name": my_table})
```
