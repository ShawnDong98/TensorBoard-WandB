# Hyperparameter Tuning

使用Weights & Biases Sweeps自动进行超参数优化，并探索可能模型的空间。

# Benefits of using W&B Sweeps

- **Quick setup**: 只需几行代码即可开始。您可以在数十台机器上 launch a sweep，这就像在笔记本电脑上开始 sweep 一样简单。
- **Transparent**: 我们引用了我们正在使用的所有算法，我们的代码是开源的。
- **Powerful**: 我们的 sweeps 是完全可定制和可配置的。

# Common Use Cases

- **Explore**: 高效地采样超参数组合的空间，以发现有前途的区域，并建立关于模型的直觉。
- **Optimize**: 使用 sweeps 找到一组具有最佳性能的超参数。

# Approach

- **Add** `wandb`: 在Python脚本中，添加几行代码来记录超参数和脚本中的输出指标。
- **Write config**: 定义要扫描的变量和范围。选择搜索策略——我们支持 grid、random 和 Bayesian search，以及 early stopping 等更快迭代的技术。在此处查看一些配置示例。
- **Initialize sweep**: 启动 sweep server。我们 host 这个 central controller，并在执行 sweep 的代理之间进行协调
- **Launch agent(s)**: 在您想要用于训练 sweeps 模型的每台机器上运行单行命令。代理询问中央 sweep 服务器下一步要尝试哪些超参数，然后执行运行。
- **Visualize results**: 打开我们的实时 dashboard ，在一个中心位置查看您的所有结果。
